```python
import pandas as pd
```

```python
df = pd.read_csv('finantialStatements(ms&tesla).csv')
df.info()
```

```python
df['Total_Revenue'] = df['Total_Revenue'].str.replace(',', '')  # Remove commas
df['Total_Revenue'] = df['Total_Revenue'].astype(float)         # Convert to float
```

```python
df['Total_assets'] = df['Total_assets'].str.replace(',', '')  # Remove commas
df['Total_assets'] = df['Total_assets'].astype(float)         # Convert to float
```

```python
df['Net_Income'] = df['Net_Income'].str.replace(',', '')  # Remove commas
df['Net_Income'] = df['Net_Income'].astype(float)         # Convert to float
```

```python
df['Total_Libilities'] = df['Total_Libilities'].str.replace(',', '')  # Remove commas
df['Total_Libilities'] = df['Total_Libilities'].astype(float)         # Convert to float
```

```python
df['Cash_flow_opera'] = df['Cash_flow_opera'].str.replace(',', '')  # Remove commas
df['Cash_flow_opera'] = df['Cash_flow_opera'].astype(float)         # Convert to float
```

```python
df.head(7)
```

```python
df=df.sort_values(['CompanyName','Years'],ascending=[True,False])
df['RevenueGrowth(&)'] = df.groupby(df['CompanyName'])['Total_Revenue'].pct_change()*100
df
```

```python
df['NetIncomeGrowth(%)'] = df.groupby(['CompanyName'])['Net_Income'].pct_change() * 100
df
```

## Finding Insights 
Financial Summary: Microsoft vs Tesla

### Microsoft:
Revenue declined consistently: -13.5% (2023), -6.4% (2022).

Net Income fell -17.9% in 2023, with slight recovery +0.6% in 2022.

Insight: Microsoft shows a downward trend in both revenue and profit, possibly due to rising costs or slowed growth.

### Tesla
Revenue dipped slightly: -0.9% (2023), and more in 2022: -15.8%.

Net Income surged +109.3% in 2023, then dropped -15.9% in 2022.

Insight: Tesla improved profitability in 2023 despite revenue dip, indicating stronger operational efficiency.
